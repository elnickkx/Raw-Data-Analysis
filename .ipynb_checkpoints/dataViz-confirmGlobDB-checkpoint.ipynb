```python
# evaluating the percentage of missing Data for clear dept on what sort data % is unknown

def missing_dataPercentage(dataframe):
    nullValues = dataframe.isnull().sum().sort_values(ascending = False)
    nullPercentage = round(dataframe.isnull().sum().sort_values(ascending = False)/len(dataframe) * 100, 2)
    
    return pd.concat([nullValues, nullPercentage], axis=1, keys=["nullTotal", "nullPercentage"])

```

```python

existValue = deathStateChange.loc[:,].sum()
#print(type(existValue))
print(existValue)
deathStateChange
```

```python
# Series to string results

def seriesToDataframe(seriesData):
    return list(np.reshape(np.asarray(seriesData), (1, np.size(seriesData)))[0])  

# str(...))[1:-1] to be used 
# when converting directly to String value, for more promising visualisation, 
# .. on idea whether i'm on right phase !!!!


seriesToDataframe(existValue)[1:]
valueSum = pd.DataFrame(seriesToDataframe(existValue)[1:])
print(valueSum)

deathStateChange["DeathCount"] = valueSum
deathStateChange
```

```python

missing_dataPercentage(deathStateChange)
```

```python

deathStateChange.loc[:, :].value_counts()
# Percentage count for unique region report

def percentValueCounts(dataframe, feature):
    countPercent = pd.DataFrame(round(dataframe.loc[:,feature].value_counts(dropna=False, normalize=True)*100,2))
    countTotal = pd.DataFrame(dataframe.loc[:,feature].value_counts(dropna=False))
    
    countTotal.columns = ["Total"]
    countPercent.columns = ['Percent']
    return pd.concat([countTotal, countPercent], axis = 1)

```

```python

plotterValue = percentValueCounts(deathStateChange, "Country/Region")
print(plotterValue)
print(list(plotterValue))
# visualising the CountryWide death cases

#plotterValue.set_index("Percent").plot(figsize=(13,9), grid=True)
plt.figure(figsize=(9, 6))
plt.scatter(
    plotterValue['Percent'],
    plotterValue['Total'],
    c='black'
)
plt.xlabel("Percent")
plt.ylabel("Total")
plt.show()

sns.catplot(x="Percent", y="Total" , data=plotterValue,
            kind="swarm")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

X = plotterValue['Percent'].values.reshape(-1,1)
y = plotterValue['Total'].values.reshape(-1,1)

reg = LinearRegression()
reg.fit(X, y)

predictions = reg.predict(X)
plt.scatter(
    plotterValue['Percent'],
    plotterValue['Total'],
    c='black'
)

plt.plot(
    plotterValue['Total'],
    predictions,
    c='blue',
    linewidth=1
)

plt.xlabel("Percent")
plt.ylabel("Total")
plt.show()
```
